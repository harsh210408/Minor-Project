## **i) Project 1 : Student Marks Analyzer (Using NumPy)**

#####**--> Problem Statement :** Developing a Python program that accepts the marks of multiple students and analyzes the marks using Python and NumPy array.




```python
# Importing NumPy Library
import numpy as np

# Defining a function to assign grades
def grade(average_marks):

  if (average_marks > 85):
    return("Grade A")
  elif (70 <= average_marks <= 85):
    return("Grade B")
  elif (50 <= average_marks <= 69):
    return("Grade C")
  else:
    return("Fail")

# Main Program
number_students = int(input("Enter Number of Students : "))
student_marks = []

print('\n')

# Entering marks as input
for i in range(number_students):
  marks = float(input(f"Enter Marks for Student {i + 1} : "))
  student_marks.append(marks)

# Converting list to NumPy array
marks_array = np.array(student_marks)

total_marks = np.sum(marks_array)
average_marks = np.mean(marks_array)
highest_marks = np.max(marks_array)
lowest_marks = np.min(marks_array)

# Calling a function
result = grade(average_marks)

# Displaying all parameters
print("\n============Students Marks Analyzer============\n")
print(f"Total Marks   : {total_marks}")
print(f"Average Marks : {average_marks}")
print(f"Highest Marks : {highest_marks}")
print(f"Lowest Marks  : {lowest_marks}")
print(f"Result        : {result}")
```

####**Logic**
**Step 1:** Importing Numpy library to use it in our code.

**Step 2:** Defining a user-defined funtion grade() to assign grades to the students, using if - elif - else statements.

**Step 3:** Taking input from the user for how many students we have to enter the marks.  

**Step 4:** Creating an empty list to store the marks of the students.

**Step 5:** By using 'for' loop we take the marks as input from the user and storing them in a list.

**Step 6:** Now, converting our list of marks into a Numpy array.

**Step 7:** Calculate (using Numpy library):

         - Total Marks
         - Average Marks
         - Highest Marks
         - Lowest Marks

**Step 8:** Pass the average marks to the grade() function.

**Step 9:** Finally displaying:
        
         - Total Marks using np.sum()
         - Average Marks np.mean()
         - Highest Marks np.max()
         - Lowest Marks np.min()
         - Final Result (Grades)

**Step 10:** Stop the program.

## **ii) Project 2 : Daily Expense Tracker (Using NumPy)**

**--> Problem Statement :** Developing a Python program that records daily expenses of the user and analyzes the spending and display the total expenses, average expenses and the day with the highest expense.

```python
# Importing NumPy Library
import numpy as np

# Defining a function to calculate total expense, average expense and highest expense day
def expense(expenses_array):
  total = np.sum(expenses_array)
  average = np.mean(expenses_array)

  highest = expenses_array[0]
  day = 1
  for i in range(len(expenses_array)):
    if(expenses_array[i] > highest):
      highest = expenses_array[i]
      day = i + 1

  return total, average, day

# Main Program
days = int(input("Enter number of days : "))
daily_expenses = []

print('\n')

# Entering daily expenses as input from the user
for i in range(days):
  money_spent = int(input(f"Enter expense for Day {i + 1} : Rs."))
  daily_expenses.append(money_spent)

# Convering list to NumPy array
expenses_array = np.array(daily_expenses)

# Calling a function and returning the values
total_expense, average_expense, highest_expense_day = expense(expenses_array)

# Displaying all the parameters
print("\n============Daily Expense Summary============\n")
print(f"Total Expense       : Rs.{total_expense}")
print(f"Average Expense     : Rs.{average_expense}")
print(f"Highest Expense Day : Day {highest_expense_day}")

# Displaying the warning if average expense exceeds a certain limit
limit = 1000
if(average_expense > limit):
  print('\n', "==================WARNING==================", '\n', "Your Average Expense Exceeds the Given Limit !!")
```

####**Logic**
**Step 1:** Importing Numpy library to use it in our code.

**Step 2:** Define a user - defined function expense() which calculates:
        
           - Total Expense using np.sum()
           - Average Expense using np.mean()
           - Highest Expense Day using 'for' loop

**Step 3:** After calculating, return all the values.

**Step 4:** Ask the user to enter the number of days.

**Step 5:** Creates an empty list to store the daily expenses.

**Step 6:** Using 'for' loop and taking daily expense as input from the user and storing them in the list.

**Step 7:** Convert list into the NumPy array.

**Step 8:** Calling the user - defined function expense() to return the values.

**Step 9:** Displaying all the returned values:

            - Total Expense
            - Average Expense
            - Highest Expense Day

**Step 10:** Now, at last we check that if the average expense exceeds the given limit.

**Step 11:** For this code, we have taken limit = 1000, so if average expense exceeds 1000, it will display the warning message to the user.

**Step 12:** End the Program.
